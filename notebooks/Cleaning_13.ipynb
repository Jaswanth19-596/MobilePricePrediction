#### 2. MainCamera_Cameraspecs


```python
import re

def extract_camera_mp(camera):

  if pd.isna(camera):
    return 0

  pattern = r'(\d+(?:\.\d+)?)\s*MP'

  matches = re.findall(pattern, camera)

  if len(matches) > 0:
    return int(float(matches[0]))

  return 0
```

```python
df['maincamera_mp'] = df['MainCamera_Cameraspecs'].apply(extract_camera_mp)
```

#### 3. SelfieCamera_Cameraspecs


```python
import re
import math

def extract_camera_mp(camera):

  if pd.isna(camera):
    return 0

  pattern = r'(\d+(?:\.\d+)?)\s*MP'

  matches = re.findall(pattern, camera)

  if len(matches) > 0:
    return math.ceil(float(matches[0]))

  camera = camera.lower()
  if 'main' in camera or 'no' in camera:
    return 0

  if 'vga' in camera or 'yes' in camera:
    return 1


  return 0
```

```python
df['selfiecamera_mp'] = df['SelfieCamera_Cameraspecs'].apply(extract_camera_mp)
```

#### 4. COMMS_WLAN

- Creating new column => Latest_wifi_version.
- Extracts the latest wifi version and gives it a integer value.
- Latest wifi version => Latest phones with latest hardware => Higher price.

'7': 5,
'6e': 4,
'6': 3,
'ac': 2,
'n': 1,
'g': 1,
'b': 1,
'a': 1


```python
def extract_latest_wifi_version(wifi):

  if pd.isna(wifi):
    return wifi

  version_score = {
        '7': 5,
        '6e': 4,
        '6': 3,
        'ac': 2,
        'n': 1,
        'g': 1,
        'b': 1,
        'a': 1
    }

  wifi = wifi.lower()

  version = wifi.split(',')[0].split(' ')[-1].split('/')[-1]

  try:
    max_version = version_score[version]
    return max_version

  except:
    return 0
```

```python
df['latest_wifi_version'] = df['COMMS_WLAN'].apply(extract_latest_wifi_version)
```

#### 5. COMMS_Bluetooth


```python
allfeatures = []
for features in df['COMMS_Bluetooth'].value_counts().index.str.split(','):
  allfeatures.extend(features)
```

```python
def extract_bluetooth_version(bluetooth):

  if pd.isna(bluetooth):
    return bluetooth

  version = bluetooth.split(',')[0]

  floats = re.findall(r'\d+\.\d+', version)

  if version == 'Yes':
    return 0

  if len(floats) == 0:
    return 0

  return float(floats[0])
```

```python
df['Bluetooth_version'] = df['COMMS_Bluetooth'].apply(extract_bluetooth_version)
```

#### 6. COMMS_NFC


```python
def has_nfc(nfc):
  if pd.isna(nfc):
    return nfc

  nfc = nfc.lower()

  if 'yes' in nfc:
    return 1

  return 0
```

```python
df['has_nfc'] = df['COMMS_NFC'].apply(has_nfc)
```

#### 8. COMMS_USB


```python
def get_usb_type(usb):
  if pd.isna(usb):
    return usb

  usb = usb.lower()

  if 'type-c' in usb:
    return 'Type-C'
  if 'microusb' in usb:
    return 'Type-B'
  if 'lightning' in usb:
    return 'Lightning'

  return 'Other'
```

```python
df['USB_Type'] = df['COMMS_USB'].apply(get_usb_type)
```
